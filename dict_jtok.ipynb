```
f_j=open("Zyukugo-xenkwan-ki/Segsyoxafu_zihom.txt")
```

```
lines_j = f_j.readlines() # 1行毎にファイル終端まで全て読む(改行文字も含まれる)
f_j.close()
```

```
lines_j[1:]
```

```
d_j={}
```

```
for line in lines_j[1:]:
  line_splited=line.split(" ")
  result=line_splited[1].split("/")
  om=line_splited[0]
  for kanji in result[1:len(result)-1]:
    d_j[kanji]=om
```

```
d_j
```

```
d_ktoj={}
d_jtok={}
```

```
      d_jtok[keyjtok]=d_jtok[keyjtok]+key_k
      d_ktoj[keyktoj]=d_ktoj[keyktoj]+key_k
```

```
for key_k, value_k in d_k.items():
  if key_k in d_j:
    keyjtok=d_j[key_k]+":"+value_k
    keyktoj=value_k+":"+d_j[key_k]
    if keyjtok in d_jtok:
      d_jtok[keyjtok]=d_jtok[keyjtok]+key_k
      d_ktoj[keyktoj]=d_ktoj[keyktoj]+key_k
    else:
      d_jtok[keyjtok]=key_k
      d_ktoj[keyktoj]=key_k
```
